# Classes

```python
class Bird():
    def __init__(self):
        self.x=100
        self.y=400
        self.w=46
        self.h=34
        self.angle=0
        self.acel=2
        self.vel=0
        self.time_ini=0
        self.time_trans=0
        
        
    def tiempo_ini(self):
        self.time_ini=time.time()
        
    def tiempo_dif(self):
        self.time_trans= (time.time()-self.time_ini)
    
    def gravedad(self):
#        self.vel=self.vel+(self.acel*self.time_trans)
#        self.y=self.y+(self.vel*self.time_trans)
        self.vel=self.vel-(self.acel*self.time_trans)
        self.y=self.y-self.vel
        
    def jump(self):
        self.tiempo_ini()
        self.vel=8
#        if self.jumpcount>=0:
#            self.y=self.y-(self.jumpcount)**2
#            self.jumpcount-=1
#            
#        elif self.jumpcount <0 and self.jumpcount>=-7:
#            self.y=self.y+(self.jumpcount)**2
#            self.jumpcount-=1
#        
#        else:
#            self.isjump=False
#            self.jumpcount=7
    
    def drawBird(self,im):
#        pygame.draw.rect(displaySurface,red,(self.x,self.y,self.w,self.h))
        displaySurface.blit(im, [self.x-10,self.y-3])
    
class Pipe():
    def __init__(self):
        self.x=600
        self.y=0
        self.w=150
        self.h=random.randint(100,500)
        self.gap=150
    
    def moverse(self):
        self.x-=5
    
    def drawPipeUpper(self,im):
        displaySurface.blit(im, [self.x,self.h-500])
#        pygame.draw.rect(displaySurface,red,(self.x,self.h-500,self.w,500))
    
    def drawPipeLower(self,im):
        displaySurface.blit(im, [self.x,self.h+self.gap])
#        pygame.draw.rect(displaySurface,red,(self.x,self.h+self.gap,self.w,500))
      
    
class Fondo():
    def __init__(self):
        self.x=0
        self.y=0
        self.w=880
    
    def moverse(self):
        self.x-=3
    
    def drawBG(self,im,x,y):
        displaySurface.blit(im, [x,y])
```

# Game

```python
def drawTime(time_pas):
    timeSurf=font.render("TIME: %s" %(time_pas),True,white)
    timeWrite=timeSurf.get_rect()
    timeWrite.center=(300,15)
    displaySurface.blit(timeSurf,timeWrite)
    
def drawText(string,size,location,color):
    font_var = pygame.font.Font('freesansbold.ttf', size) 
    textSurf=font_var.render(string,True,color)
    textWrite=textSurf.get_rect()
    textWrite.center=location
    displaySurface.blit(textSurf,textWrite)
```

```python
fps=30

create_pipe = pygame.USEREVENT + 1
pygame.time.set_timer(create_pipe, 2500)

def main():
    global displaySurface, clock, font, start_time
    
    pygame.init()
    displaySurface=pygame.display.set_mode((600, 800))
    
    pygame.display.set_caption('FLAPPY BIRD')
    clock = pygame.time.Clock()
    font = pygame.font.Font('freesansbold.ttf', 18)
    
    player=Bird()
    pipes=[]
    
    bg_image = pygame.image.load("bg_resize.png").convert()
    bird_image = pygame.image.load("bird_resize.png").convert_alpha()
    pipe_lower = pygame.image.load("pipe_lower.png").convert_alpha()
    pipe_upper = pygame.image.load("pipe_upper.png").convert_alpha()
    fondo=Fondo()
    
    score_play=0
    
    run=True
    empezar=False
    
    while run:
        if empezar==False:
        
            for event in pygame.event.get():
                if event.type==KEYDOWN:
                    empezar=True
                
                else:
                    pass
    
        else:
            if player.time_ini==0:
                player.tiempo_ini()
                start_time=pygame.time.get_ticks()
            else:
                pass
            
            
            tiempo_ini=round(((pygame.time.get_ticks()-start_time)/1000),1)
            
            for event in pygame.event.get():
                if event.type==QUIT:
                    run=False
                    
                elif (event.type==KEYDOWN):
                    keys = pygame.key.get_pressed()
                    
                    if keys[pygame.K_SPACE]:
                        player.jump()
#                        player.isjump=True
                        
                elif (event.type==create_pipe):
                    pipes.append(Pipe())
            
            
            if (fondo.x<=-880):
                fondo.x=0
            
            fondo.drawBG(bg_image,fondo.x,fondo.y)
            fondo.drawBG(bg_image,fondo.x+880,fondo.y)
            fondo.moverse()
            
            for i in pipes:
                i.moverse()
                i.drawPipeUpper(pipe_upper)
                i.drawPipeLower(pipe_lower)
            
            player.tiempo_dif()
            
            print(player.vel,player.y)
            
            player.gravedad()
            
            player.drawBird(bird_image)
            
            
        
            drawTime(tiempo_ini)
            
            pygame.display.update()
            clock.tick(fps)
```

```python
main()
```
